## Drawing frame and image drawing

```python
h, w = 1080, 1920
bg_colour = (255, 255, 255)
bg_image = np.dot(np.ones((h,w,3), dtype='uint8'), np.diag(np.asarray((bg_colour), dtype='uint8')))
image0 = Image.fromarray(bg_image)
draw = ImageDraw.Draw(image0)
```
